```python
import numpy as np
import matplotlib.pyplot as plt
import os.path
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from DBSCAN_Clustering import dbscan_clustering_sklearn
```

```python
X = np.load(os.path.join(os.path.abspath("résultats (1ère base)"),"Features_matrix_1(scaled).npy"))
print(X.shape)
N = X.shape[0]
```

```python
# minPts_array = np.arange(4,5,1)
n_eps = 5
silhouette_score_array = np.zeros(n_eps)
calinski_harabasz_score_array = np.zeros(n_eps)
```

## Distance euclidienne

```python
metric_ = "euclidean"
minPts = 4
eps_min = 7.54
eps_max = 7.64
eps_array = np.arange(eps_min, eps_max, (eps_max-eps_min)/n_eps)

for i in range(n_eps):
    classes = dbscan_clustering_sklearn(X,eps_array[i],minPts,metric_)
    n_clusters = np.max(classes)+1
    print(n_clusters)
    silhouette_score_array[i] = silhouette_score(X,classes)
    calinski_harabasz_score_array[i] = calinski_harabasz_score(X,classes)*(n_clusters-1)/(N-n_clusters)
```

```python
np.save(os.path.join(os.path.abspath("résultats (1ère base)"),"Silhouette_score_dbscan(euclidean)"), silhouette_score_array)
```

```python
np.save(os.path.join(os.path.abspath("résultats (1ère base)"),"Calinski_harabasz_dbscan(euclidean)"), calinski_harabasz_score_array)
```

```python
plt.figure(figsize=(20,6))
ax1 = plt.subplot(1,2,1)
ax1.plot(eps_array, silhouette_score_array)
ax1.set_title("Silhouette_score")
ax1.set_xlabel("eps")
ax2 = plt.subplot(1,2,2)
ax2.plot(eps_array, calinski_harabasz_score_array)
ax2.set_title("Calinski_harabasz_score ")
ax2.set_xlabel("eps")
plt.show()
```
