```python
import numpy as np
import matplotlib.pyplot as plt
import os.path
from sklearn.metrics import silhouette_score, calinski_harabaz_score
from KMeans_Clustering import kmeans_pp, kmeans_clustering_sklearn
```

```python
X = np.load(os.path.join(os.path.abspath("résultats"),"Features_matrix_2(scaled).npy"))
print(X.shape)
```

```python
n_clusters_max = 20
n_clusters_min = 4
n_clusters_array = np.arange(n_clusters_max-1)+ n_clusters_min
lc = n_clusters_array.shape[0]
silhouette_score_array = np.zeros(lc)
calinski_harabaz_score_array = np.zeros(lc)
```

## Distance euclidienne

```python
metric_ = "euclidean"
n_init = 3

for k in n_clusters_array:
#     classes = kmeans_clustering_sklearn(X,k)
    classes = kmeans_pp(X,k,n_init,metric_)
    silhouette_score_array[k-n_clusters_min] = silhouette_score(X,classes)
    calinski_harabaz_score_array[k-n_clusters_min] = calinski_harabaz_score(X,classes)
```

```python
np.save(os.path.join(os.path.abspath("résultats"),"Silhouette_score_kmeans_2(euclidean)"), silhouette_score_array)
```

```python
plt.figure(figsize=(20,6))
ax1 = plt.subplot(1,2,1)
ax1.plot(n_clusters_array, silhouette_score_array)
ax1.set_title("Silhouette_score vs number of clusters")
ax2 = plt.subplot(1,2,2)
ax2.plot(n_clusters_array, calinski_harabaz_score_array)
ax2.set_title("Calinski_harabasz_score vs number of clusters")
plt.show()
```

```python
silhouette_euclidean = np.load(os.path.join(os.path.abspath("résultats"),"Silhouette_score_kmeans_2(euclidean).npy"))
silhouette_manhattan = np.load(os.path.join(os.path.abspath("résultats"),"Silhouette_score_kmeans_2(manhattan).npy"))
silhouette_cosine = np.load(os.path.join(os.path.abspath("résultats"),"Silhouette_score_kmeans_2(cosine).npy"))
silhouette_chebyshev = np.load(os.path.join(os.path.abspath("résultats"),"Silhouette_score_kmeans_2(chebyshev).npy"))

distances = ["euclidean", "manhattan", "cosine", "chebyshev"]
silhouette = [np.max(silhouette_euclidean), np.max(silhouette_manhattan), np.max(silhouette_cosine), np.max(silhouette_chebyshev)]
k_all = np.array([np.argmax(silhouette_euclidean), np.argmax(silhouette_manhattan), \
                  np.argmax(silhouette_cosine), np.argmax(silhouette_chebyshev)]) + 4

print('Number of clusters: ')
print('Euclidean: ', k_all[0])
print('Manhattan: ', k_all[1])
print('Cosine: ', k_all[2])
print('Chebyshev: ', k_all[3])
plt.figure()
plt.barh(distances, silhouette)
plt.title("Comparison between distances (KMeans)")
plt.show()
```
