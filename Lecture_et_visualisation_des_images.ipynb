Parfois, on a besoin de changer la résolution d'une image, notamment pour diminuer le temps de calcul. On utilise alors un outil d'openCV qui permet de le faire en conservant le ratio longueur/largeur de l'image pour éviter d'avoir une distorsion.

```python
from Image_reading import *
from Visualisation import *
import cv2 as cv
if __name__ == '__main__':
    X=read_image_2(1,0)
    nx=X.shape[0]
    ny=X.shape[1]
    r = 100.0 / ny
    dim = (100, int(nx * r))
    X_resized = cv.resize(X, dim, interpolation = cv.INTER_AREA)
    print (X.shape)
    plot_image (X)
    print(X_resized.shape)
    plot_image(X_resized)
```
