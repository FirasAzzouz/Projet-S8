```python
import numpy as np
import matplotlib.pyplot as plt
import os.path
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from Spectral_Clustering import spectralClustering_rbf
```

```python
X = np.load(os.path.join(os.path.abspath("résultats (1ère base)"),"Features_matrix_1(scaled).npy"))
print(X.shape)
```

```python
n_clusters_max = 20
n_clusters_min = 4
n_clusters_array = np.arange(n_clusters_max-1)+ n_clusters_min
lc = n_clusters_array.shape[0]
silhouette_score_array = np.zeros(lc)
calinski_harabasz_score_array = np.zeros(lc)
```

## Distance euclidienne

```python
metric_ = "euclidean"
gamma_ = 1e-4

for k in n_clusters_array:
    classes = spectralClustering_rbf(X,k,metric_,gamma_)
    silhouette_score_array[k-n_clusters_min] = silhouette_score(X,classes)
    calinski_harabasz_score_array[k-n_clusters_min] = calinski_harabasz_score(X,classes)
```

```python
np.save(os.path.join(os.path.abspath("résultats (1ère base)"),"Silhouette_score_spectral(euclidean)"), silhouette_score_array)
```

```python
plt.figure(figsize=(20,6))
ax1 = plt.subplot(1,2,1)
ax1.plot(n_clusters_array, silhouette_score_array)
ax1.set_title("Silhouette_score vs number of clusters")
ax2 = plt.subplot(1,2,2)
ax2.plot(n_clusters_array, calinski_harabaz_score_array)
ax2.set_title("Calinski_harabasz_score vs number of clusters")
plt.show()
```
